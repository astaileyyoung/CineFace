```python
from pathlib import Path

import pandas as pd
from tqdm import tqdm
```

```python
faces_dir = Path('../data/faces')
subdirs = [x for x in faces_dir.iterdir()]
```

```python
for subdir in tqdm(subdirs):
    files = [x for x in subdir.iterdir()]
    for file in tqdm(files, leave=False):
        df = pd.read_csv(str(file), index_col=0)
        df = df.assign(pct_of_frame=df.apply(lambda row: (row['x2'] - row['x1']) * (row['y2'] - row['y1']), axis=1))
        df.to_csv(file)
```

```python
files = [x for x in subdirs[1].iterdir()]
df = pd.read_csv(str(files[0]), index_col=0)
df.head()
```

```python
df['pct_of_frame'] = df.apply(lambda row: (row['x2'] - row['x1']) * (row['y2'] - row['y1']), axis=1)

df['face_width'] = df.apply(lambda x: (x['x2'] * x['img_width']) - (x['x1'] * x['img_width']), axis=1)
df['face_height'] = df.apply(lambda x: (x['y2'] * x['img_height']) - (x['y1'] * x['img_height']), axis=1)
df['face_area'] = df['face_width'] * df['face_height']
df['face_pct_of_frame'] = df.apply(lambda x: x['face_area']/ (x['img_width'] * x['img_height']), axis=1)
df.head()
```
