```python
import cv2 
import pandas as pd
import IPython
```

```python
base = pd.read_csv('/home/amos/shining_python.csv')
print(base.columns)
base
```

```python
c = pd.read_csv('/home/amos/shining_c++.csv', index_col=0)
c
```

```python
base[~base['frame_num'].isin(c['frame_num'])]
```

```python
merged = pd.merge(base, c, on='frame_num', suffixes=('_base', '_c'))
result = merged[merged['face_num_base'] > merged['face_num_c']]
result
```

```python
src = '/home/amos/datasets/sample_videos/shining_bat.mp4'
cap = cv2.VideoCapture(src)
framecount = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
for idx, row in result.iterrows():
    cap.set(cv2.CAP_PROP_POS_FRAMES, row['frame_num'])
    ret, frame = cap.read()
    base_temp = base[base['frame_num'] == row['frame_num']]
    for i, r in base_temp.iterrows():
        cv2.rectangle(frame, (r['x1'], r['y1']), (r['x2'], r['y2']), (0, 255, 0))
    
    c_temp = c[c['frame_num'] == row['frame_num']]
    for i, r in c_temp.iterrows():
        cv2.rectangle(frame, (r['x1'], r['y1']), (r['x2'], r['y2']), (0, 0, 255))

    h, w = frame.shape[:2]
    resized = cv2.resize(frame, (int(w/2), int(h/2)))
    _, ret = cv2.imencode('.jpg', resized)
    i = IPython.display.Image(data=ret)
    IPython.display.display(i)

print('-----------------------C NOT IN BASE---------------------------')
temp = c[~c['frame_num'].isin(base['frame_num'])]
for idx, row in temp.iterrows():
    cap.set(cv2.CAP_PROP_POS_FRAMES, row['frame_num'])
    ret, frame = cap.read()
    cv2.rectangle(frame, (row['x1'], row['y1']), (row['x2'], row['y2']), (0, 0, 255))
    h, w = frame.shape[:2]
    resized = cv2.resize(frame, (int(w/2), int(h/2)))
    _, ret = cv2.imencode('.jpg', resized)
    i = IPython.display.Image(data=ret)
    IPython.display.display(i)

print('-----------------------BASE NOT IN C---------------------------')
temp = base[~base['frame_num'].isin(c['frame_num'])]
for idx, row in temp.iterrows():
    cap.set(cv2.CAP_PROP_POS_FRAMES, row['frame_num'])
    ret, frame = cap.read()
    cv2.rectangle(frame, (row['x1'], row['y1']), (row['x2'], row['y2']), (0, 255, 0))
    h, w = frame.shape[:2]
    resized = cv2.resize(frame, (int(w/2), int(h/2)))
    _, ret = cv2.imencode('.jpg', resized)
    i = IPython.display.Image(data=ret)
    IPython.display.display(i)
```
