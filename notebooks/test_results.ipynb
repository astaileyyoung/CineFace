```python
import cv2
import IPython
import pandas as pd
```

```python
def draw_face(img, row):
    x1 = int(row['x1'] * row['img_width'])
    y1 = int(row['y1'] * row['img_height'])
    x2 = int(row['x2'] * row['img_width'])
    y2 = int(row['y2'] * row['img_height'])
    cv2.rectangle(img, (x1, y1), (x2, y2), (0, 0, 255), 1)
```

```python
def show_image(image):
    _, ret = cv2.imencode('.jpg', image)
    i = IPython.display.Image(data=ret)
    IPython.display.display(i)
```

```python
df = pd.read_csv('../data/shining.csv', index_col=0)
df.head()
```

```python
cap = cv2.VideoCapture('../data/videos/shining_bat.mp4')
frame_num = df.at[10, 'frame_num']
print(frame_num)
cap.set(cv2.CAP_PROP_POS_FRAMES, frame_num)
```

```python
ret, frame = cap.read()
show_image(frame)
```

```python
draw_face(frame, df.iloc[10])
```
