```python
import cv2
import IPython
```

```python
def show_image(image):
    _, ret = cv2.imencode('.jpg', image)
    i = IPython.display.Image(data=ret)
    IPython.display.display(i)
```

```python
img = cv2.imread('./images/img_1.jpg')
show_image(img)
```

```python
height, width = img.shape[:2]
height_border = int(0.5 * height)
width_border = int(0.5 * width)
img_border = cv2.copyMakeBorder(
    img,
    height_border,
    height_border,
    width_border,
    width_border,
    cv2.BORDER_CONSTANT,
    value=[0, 0, 0]
)
show_image(img_border)
```
