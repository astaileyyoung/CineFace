```python
import cv2
import IPython
from batch_face import RetinaFace
```

```python
def show_image(image):
    _, ret = cv2.imencode('.jpg', image)
    i = IPython.display.Image(data=ret)
    IPython.display.display(i)
```

```python
def draw_faces(image, faces):
    for box, landmarks, confidence in faces:
        # バウンディングボックス
        x1, y1, x2, y2 = [int(x) for x in box]
        color = (0, 0, 255)
        thickness = 2
        cv2.rectangle(image, (x1, y1), (x2, y2), color, thickness, cv2.LINE_AA)
    show_image(image)
```

```python
def draw_landmarks(image, landmarks):
    for num, landmark in enumerate(landmarks):
        l = tuple(int(x) for x in tuple(landmark))
        cv2.circle(image, l, radius=10, color=(0, 0, 255), thickness=-1)
        cv2.putText(image, str(num), (l[0] + 10, l[1]), cv2.FONT_HERSHEY_PLAIN, 5, (0, 0, 255), 2)
    show_image(image)
```

```python
img = cv2.imread('../test_images/Billions.S02E07.1080p.BluRay.x265-RARBG_39816.png')
```

```python
det = RetinaFace()
```

```python
faces = det(img)
box, landmarks, confidence = faces[0]
```

```python
draw_faces(img, faces)
```
