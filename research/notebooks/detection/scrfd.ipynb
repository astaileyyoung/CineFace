```python
from pathlib import Path 

import cv2
import IPython
import numpy as np
import insightface
from insightface.app import FaceAnalysis
from insightface.data import get_image
```

```python
def show_image(image):
    _, ret = cv2.imencode('.jpg', image)
    i = IPython.display.Image(data=ret)
    IPython.display.display(i)
```

```python
def draw_faces(image, faces):
    img = image.copy()
    for face in faces:
        bbox = face['bbox']
        x1, y1, x2, y2 = [int(x) for x in list(bbox)]
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 0, 255), 1)
        for num, landmark in enumerate(face['kps']):
            cv2.circle(img, (int(landmark[0]), int(landmark[1])), 5, (0, 0, 255), -1)
            cv2.putText(img, str(num), (int(landmark[0]), int(landmark[1])), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 255, 0), 3, cv2.LINE_AA)
    show_image(img)
```

```python
def detect_from_name(filename):
    fp = Path('../test_images/').joinpath(filename)
    img = cv2.imread(str(fp))
    app = FaceAnalysis(allowed_modules=['detection'], name='buffalo_l')
    app.prepare(ctx_id=0, det_size=(640, 640))
    faces = app.get(img)
    draw_faces(img, faces)
    
```
