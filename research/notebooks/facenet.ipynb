I'm checking whether the low accuracy has to do with the detection or the encoding by comparing the same encoding against the RetinaFace detections.

```python
yunet_labeled = pd.read_csv('../../data/yunet_facenet_test_labeled.csv', index_col=0)
yunet_labeled = yunet_labeled.reset_index(drop=True)
yunet_labeled.head()
```

```python
retina_labled = label_df.copy()
cap = cv2.VideoCapture(label_df.at[0, 'filepath'])
for idx, row in tqdm(label_df.iterrows(), total=label_df.shape[0]):
    cap.set(cv2.CAP_PROP_POS_FRAMES, row['frame_num'])
    ret, frame = cap.read()
    x1 = int(row['x1'] * row['img_width'])
    y1 = int(row['y1'] * row['img_height'])
    x2 = int(row['x2'] * row['img_width'])
    y2 = int(row['y2'] * row['img_height'])
    face = frame[y1:y2, x1:x2]
    f = preprocessing.normalize_input(preprocessing.resize_image(face, (160, 160)), normalization='Facenet')
    encoding = model(f)[0]._numpy().tolist()
    retina_labled.at[idx, 'encoding'] = encoding
```
