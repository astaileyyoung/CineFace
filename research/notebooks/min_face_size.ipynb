```python
from pathlib import Path

import cv2
import pandas as pd
from tqdm import tqdm 
```

```python
def pct_of_frame(row):
    x = int((row['x2'] - row['x1']) * row['img_width'])
    y = int((row['y2'] - row['y1']) * row['img_height'])
    xx = row['img_width']
    yy = row['img_height']

    pct_of_frame = (x * y)/(xx * yy)
    return round(pct_of_frame * 100, 2) 
```

```python
df = pd.read_csv('/home/amos/programs/CineFace/data/test_labeled.csv', index_col=0)
df['pct_of_frame'] = df.apply(pct_of_frame, axis=1)
df.head()
```

```python
df = df.sort_values(by='pct_of_frame', ascending=True)
df.head()
```

```python
dst = Path('/home/amos/programs/CineFace/data/images/House.S01E01')
cap = cv2.VideoCapture(df.at[0, 'filepath'])
for num, (idx, row) in enumerate(tqdm(df.iterrows(), total=df.shape[0])):
    cap.set(cv2.CAP_PROP_POS_FRAMES, row['frame_num'])
    ret, frame = cap.read()
    x1 = int(row['x1'] * row['img_width'])
    y1 = int(row['y1'] * row['img_height'])
    x2 = int(row['x2'] * row['img_width'])
    y2 = int(row['y2'] * row['img_height'])
    face = frame[y1:y2, x1:x2]
    name = f'{num}_{idx}_{frame_num}_{row["face_num"]}.png'
    fp = dst.joinpath(name)
    cv2.imwrite(str(fp), face)
```
